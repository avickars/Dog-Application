```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
dogInfo = pd.read_csv('attributes.csv', index_col=0)
```

```python
dogInfo[dogInfo['type'] == 'validation']
```

```python
dogInfo['type'] = 'train'
```

```python
train, temp = train_test_split(dogInfo.index, test_size=0.2, random_state=42)
```

```python
test, validation = train_test_split(temp, test_size=0.3, random_state=42)
```

```python
dogInfo.iloc[train, 10] = 'train'
```

```python
dogInfo.iloc[test, 10] = 'test'
```

```python
dogInfo.iloc[validation, 10] = 'validation'
```

```python
dogInfo.to_csv('attributes.csv')
```
