```python
directory = 'annotation/Annotation'
dogs_dirs_list = [x[0] for x in os.walk(directory)]
dogs_dirs_list = dogs_dirs_list[1:] # exclude the first el since it is the name of the main dir (annotation/Annotation)
dogs_dirs_list
```

```python
dogs_breeds = []
for dogs_dir in dogs_dirs_list:
    dog_breed = dogs_dir.split('/')[-1]
    dog_breed = dog_breed.split('-')[-1]
    dogs_breeds.append(dog_breed)
    
dogs_breeds
```

```python

with open("stanford_dogs_breeds.txt", "w") as output:
    output.write('\n'.join(dogs_breeds)) 
```
